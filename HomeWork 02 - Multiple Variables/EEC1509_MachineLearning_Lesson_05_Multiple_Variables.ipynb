```
import pandas as pd

# 1 - Read AmesHousing.txt into a dataframe using the tab delimiter (\t) and assig to data.
data = pd.read_csv('AmesHousing.txt', sep='\t')

# 2 - Select the first 1460 rows from data and assign to train.
train = data[:1460]

# 3 - Select the remaining rows from data and assign to test.
test = data[1460:]

# 4 - Use the dataframe.info() method to display information about each column.
data.info()

# 5 - Read the data documentation to get more familiar with each column.

# 6 - Using the data documentation, determine which column is the target column 
# we want to predict. Assign the column name as a string to target.
# >> SalePrice          2930 non-null int64
```

```
#Generate 3 scatter plots in the same column:
#The first plot should plot the Garage Area column on the x-axis against the SalePrice column on the y-axis.
#The second one should plot the Gr Liv Area column on the x-axis against the SalePrice column on the y-axis.
#The third one should plot the Overall Cond column on the x-axis against the SalePrice column on the y-axis.

import matplotlib.pyplot as plt

axis = ('Garage Area', 'Gr Liv Area', 'Overall Cond')
colors = ('red', 'blue', 'green')
axis_colors = zip(axis, colors)

for xaxis,color in axis_colors:
   
  X = data[xaxis]
  y = data['SalePrice']
  
  
  #visualize results
  plt.scatter(X.values,y.values, color=color)
  plt.xlabel(xaxis)
  plt.ylabel('SalePrice')

leg = plt.legend(axis)
leg.get_frame().set_facecolor('#FFFFFF') # Fix

plt.show()

#Read more about these 3 columns in the data documentation.
```
